# Clusterização

```python
# ==========================================
# CLUSTERIZAÇÃO + MAPA DE CALOR INTERATIVO
# ==========================================
# pip install scikit-learn joblib hdbscan folium plotly pandas openpyxl

# tenta importar hdbscan
try:
    import hdbscan
    HDBSCAN_AVAILABLE = True
except:
    HDBSCAN_AVAILABLE = False

# ============================
# CONFIGURAÇÕES GERAIS
# ============================
LAT_COL = "LATITUDE"
LON_COL = "LONGITUDE"
DATA_COL = "DATA_OCORRENCIA_BO"
HORA_COL = "HORA_OCORRENCIA_BO"

OUTPUT_DIR = "models_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================
# LEITURA DO ARQUIVO (caso queira ler de XLSX)
# ============================
# df_crimes = pd.read_excel("seu_arquivo.xlsx")

# Garantindo que LAT/LON são numéricos
df_crimes[LAT_COL] = pd.to_numeric(df_crimes[LAT_COL], errors="coerce")
df_crimes[LON_COL] = pd.to_numeric(df_crimes[LON_COL], errors="coerce")
df_crimes = df_crimes.dropna(subset=[LAT_COL, LON_COL])

# ============================
# PREPARAÇÃO DOS DADOS
# ============================
X = df_crimes[[LAT_COL, LON_COL]].to_numpy(dtype=float)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================
# TREINAMENTO E AVALIAÇÃO
# ============================
best = {"name": None, "model": None, "score": -999.0, "params": None}

param_grid = {
    "kmeans": {"n_clusters": [4, 6, 8, 10, 12]},
    "dbscan": {"eps": [0.01, 0.02, 0.03, 0.05], "min_samples": [3, 5, 10]},
}
if HDBSCAN_AVAILABLE:
    param_grid["hdbscan"] = {"min_cluster_size": [5, 10, 20], "min_samples": [None, 5]}

print("[*] Testando KMeans...")
for params in ParameterGrid(param_grid["kmeans"]):
    model = KMeans(**params, random_state=42, n_init=10)
    labels = model.fit_predict(X_scaled)
    if len(set(labels)) > 1:
        score = silhouette_score(X_scaled, labels)
        if score > best["score"]:
            best.update({"name": "kmeans", "model": make_pipeline(scaler, model), "score": score, "params": params})

print("[*] Testando DBSCAN...")
for params in ParameterGrid(param_grid["dbscan"]):
    model = DBSCAN(**params)
    labels = model.fit_predict(X_scaled)
    if len(set(labels)) > 1 and len(set(labels) - {-1}) > 1:
        score = silhouette_score(X_scaled, labels)
        if score > best["score"]:
            best.update({"name": "dbscan", "model": make_pipeline(scaler, model), "score": score, "params": params})

if HDBSCAN_AVAILABLE:
    print("[*] Testando HDBSCAN...")
    for params in ParameterGrid(param_grid["hdbscan"]):
        model = hdbscan.HDBSCAN(**params, metric="euclidean")
        labels = model.fit_predict(X_scaled)
        if len(set(labels)) > 1 and len(set(labels) - {-1}) > 1:
            score = silhouette_score(X_scaled, labels)
            if score > best["score"]:
                best.update({"name": "hdbscan", "model": make_pipeline(scaler, model), "score": score, "params": params})

print("\n===============================")
print(f"[+] Melhor modelo: {best['name']}")
print(f"[+] Silhouette Score: {best['score']:.4f}")
print(f"[+] Parâmetros: {best['params']}")

# ============================
# SALVAR O MELHOR MODELO
# ============================
model_path = os.path.join(OUTPUT_DIR, f"best_{best['name']}_score{best['score']:.4f}.joblib")
joblib.dump({"model": best["model"], "params": best["params"], "score": best["score"]}, model_path)
print(f"[+] Modelo salvo em: {model_path}")

# ============================
# APLICAR CLUSTER AO DATAFRAME
# ============================
best_model = best["model"]
df_crimes["cluster"] = best_model.named_steps[list(best_model.named_steps.keys())[-1]].fit_predict(X_scaled)

# ============================
# MAPA DE CALOR COM FOLIUM
# ============================
map_center = [df_crimes[LAT_COL].mean(), df_crimes[LON_COL].mean()]
mapa = folium.Map(location=map_center, zoom_start=13)
HeatMap(df_crimes[[LAT_COL, LON_COL]].values.tolist(), radius=10).add_to(mapa)
mapa.save(os.path.join(OUTPUT_DIR, "mapa_calor_clusters.html"))
print(f"[+] Mapa de calor salvo em: {OUTPUT_DIR}/mapa_calor_clusters.html")

# ============================
# MAPA INTERATIVO COM PLOTLY (CLUSTERS)
# ============================
fig = px.scatter_mapbox(
    df_crimes,
    lat=LAT_COL,
    lon=LON_COL,
    color="cluster",
    hover_data=["RUBRICA", "BAIRRO", "DATA_OCORRENCIA_BO", "HORA_OCORRENCIA_BO"],
    zoom=11,
    mapbox_style="carto-positron",
    title="Clusters de Criminalidade em Itapevi",
)
fig.write_html(os.path.join(OUTPUT_DIR, "mapa_clusters_interativo.html"))
print(f"[+] Mapa interativo salvo em: {OUTPUT_DIR}/mapa_clusters_interativo.html")
```
